```python
rw = pd.read_excel(io.BytesIO(uploaded.get('rfm.xlsx')))
rw
```

```python
rw.dropna(subset=['order_id','customer_id'],how ='all', inplace=True)
print(rw)
```

```python
now=dt.date(2022,3,17)
print(now)
```

```python
rw['Date']=pd.DatetimeIndex(rw['date_created_gmt']).date
print(rw)
```

```python
rw_recency=rw.groupby(by='customer_id',as_index=False)['Date'].max()
```

```python
rw_recency.columns
```

```python
rw_recency.head()
```

```python
rw['Recency']=rw_recency['Date'].apply(lambda x: (now- x).days).max()
```

```python
rw
```

```python
rw_frequency=rw.groupby(by='customer_id', as_index=False, )['order_id'].count()
rw_frequency.head()
```

```python
rw_frequency.columns=['costumer_id','Frequency']
rw_frequency.head()
```

```python
rw_monetary=rw.groupby(by='customer_id',as_index=False)['total'].sum()
rw_monetary.head()
```

```python
rw_monetary.columns=['costumer_id','Monetary']
rw_monetary.head()
```

```python
RF = pd.merge(rw_frequency , rw_recency,left_index=True,right_on=rw_frequency['costumer_id'])
RFM=pd.merge(RF,rw_monetary)
RFM.head()

```
